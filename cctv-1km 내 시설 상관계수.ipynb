```python
cctv=pd.read_csv('./cctv 인근 1km 내 씨씨티비수.csv')
```

```python
cctv.rename(columns = {"NUMPOINTS": "인근 1km 내 cctv수"}, inplace=True)
cctv=cctv.drop(['type', 'install_ym', 'lon', 'lat'],1)
```

```python
df_merge=pd.merge(cctv,보안등,on='_cctv_nm',how='outer')
df_merge=pd.merge(df_merge,지킴이집,on='_cctv_nm',how='outer')
df_merge=pd.merge(df_merge,유치원,on='_cctv_nm',how='outer')
df_merge=pd.merge(df_merge,주차장,on='_cctv_nm',how='outer')
df_merge=pd.merge(df_merge,초중고,on='_cctv_nm',how='outer')
df_merge=pd.merge(df_merge,치안유관업소,on='_cctv_nm',how='outer')

df_merge
```

```python
df_merge.columns
```

```python
df_merge.rename(columns = {'_cctv_nm':'cctv_name','인근 1km 내 cctv수':'1km_cctv_cnt', '인근 1km 내 보안등 개수':'1km_light_cnt', '인근 1km 내 지킴이집수':'1km_child_guard_cnt', '인근 1km 내 유치원수':'1km_kindergtn_cnt',
       'cctv 인근 1km 내 주차장 수':'1km_parking_cnt', '인근 1km 내 초중고수':'1km_elemidhigh_cnt', '인근 1km 내 치안유관업종':'1km_safe_inds_cnt'}, inplace=True)
```

```python
df_merge_corr=df_merge.corr()
```

```python
import seaborn as sns
import numpy as np
```

```python
mask=np.zeros_like(df_merge.corr(),dtype=np.bool)
mask[np.triu_indices_from(mask)]=True

sns.heatmap(df_merge.corr(),annot=True,mask=mask,cmap = 'Blues', square=True)
```
